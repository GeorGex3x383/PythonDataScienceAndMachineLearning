```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
df = pd.DataFrame({
    "etiquetas":['Manzanas', 'Bananas', 'Naranjas', 'Uvas'],
    "cantidad":[30, 35, 26, 33]
    })
df
```

```python
fig, axs = plt.subplots(ncols=2, nrows=1)
ax1 = axs[0]
ax2 = axs[1]
ax1.plot(df['etiquetas'], df['cantidad'])
ax2.pie(df['cantidad']);
```

```python
plt.style.use('ggplot')
fig, axs = plt.subplots(ncols=2, nrows=1)
ax1 = axs[0]
ax2 = axs[1]
ax1.plot(df['etiquetas'], df['cantidad'])
ax2.pie(df['cantidad']);
```
