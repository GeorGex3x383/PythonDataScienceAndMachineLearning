```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
df = pd.DataFrame({
    "etiquetas":['Manzanas', 'Bananas', 'Naranjas', 'Uvas'],
    "cantidad":[30, 60, 80, 20]
})
df
```

```python
plt.pie(df['cantidad']);
```

```python
plt.pie(df['cantidad'],labels=df['etiquetas']);
```

```python
plt.pie(df['cantidad'],labels=df['etiquetas'],autopct='%1.1f%%');
```

```python
plt.pie(df['cantidad'],labels=df['etiquetas'],autopct='%1.1f%%',colors=['blue','yellow','cyan','orange']);
```

```python
plt.pie(df['cantidad'],labels=df['etiquetas'],autopct='%1.1f%%',colors=['blue','yellow','cyan','orange'],shadow=True);
```

```python
plt.pie(df['cantidad'],labels=df['etiquetas'],autopct='%1.1f%%',colors=['blue','yellow','cyan','orange'],shadow=True,explode=(0,0,0,0.5));
```

```python
plt.pie(df['cantidad'],labels=df['etiquetas'],autopct='%1.1f%%',colors=['blue','yellow','cyan','orange'],shadow=True,explode=(0,0,0,0.5));
plt.title('Mis frutas');
```
