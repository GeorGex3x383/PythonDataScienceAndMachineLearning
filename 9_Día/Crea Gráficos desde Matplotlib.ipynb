```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("c:/Users/jemil/Documents/CursosYMas/CursoPythonDSyML/9_Día/Iluminacion.csv")
df
```

```python
fig, ax = plt.subplots()
hb = ax.hexbin(df['Latitud'],df['Longitud'],C=df['Luminosidad'],gridsize=50, cmap='Oranges')
ax.set_facecolor('#000000')
plt.colorbar(hb, label="Nivel de sombras")
plt.xlabel("Latitud")
plt.ylabel("Longitud")
plt.title("Luminicidad en VillaLuz");
```

```python
x = range(1, 11)  # Crear una lista de números enteros del 1 al 10
y = [i**2 for i in x]  # Calcular el cuadrado de cada número en la lista x
 
# Crear gráfico de tallo
fig, ax = plt.subplots()  # Inicializar una figura y un conjunto de subgráficos
ax.stem(x, y)  # Crear un gráfico de tallo para mostrar los datos de x e y
 
# Añadir títulos y etiquetas
ax.set_title("Relación entre Números y sus Cuadrados")  # Establecer el título del gráfico
ax.set_xlabel("Número (x)")  # Establecer la etiqueta del eje x
ax.set_ylabel("Cuadrado del número (x^2)")  # Establecer la etiqueta del eje y
```

```python
y1 = [6, 1, 4, 4, 8]
y2 = [4, 6, 3, 5, 8]
y3 = [7, 9, 9, 2, 7]
x = range(1, 6)  # Tiempos del 1 al 5
 
# Crear el gráfico apilado
fig, ax = plt.subplots()
ax.stackplot(x, y1, y2, y3, colors=['red', 'blue', 'green'], labels=['Serie y1', 'Serie y2', 'Serie y3'])
 
# Añadir leyenda
ax.legend(loc='upper left')
 
# Añadir títulos y etiquetas
ax.set_title('Distribución acumulativa de valores por serie de tiempo')
ax.set_xlabel('Tiempo')
ax.set_ylabel('Valores Acumulados')
```

```python
np.random.seed(0) # Establecer la misma semilla para la generación de datos
data1 = np.random.normal(loc=0, scale=0.5, size=100)
data2 = np.random.normal(loc=2, scale=1, size=100)
data3 = np.random.normal(loc=-2, scale=1.5, size=100)
 
# Preparar los datos para el violin plot
data = [data1, data2, data3]
 
# Crear el violin plot
fig, ax = plt.subplots()
ax.violinplot(data)
 
# Establecer las etiquetas del eje x
ax.set_xticks([1, 2, 3])
ax.set_xticklabels(['Muestra 1', 'Muestra 2', 'Muestra 3'])
 
# Añadir título al gráfico
ax.set_title('Comparación de distribuciones en Violin Plot')
```
